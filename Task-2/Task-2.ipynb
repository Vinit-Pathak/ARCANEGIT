```python
import cv2
import matplotlib.pyplot as mtp
```

```python
image = cv2.imread('../Task-2/Naruto.jpeg')
mtp.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
mtp.axis('off')
mtp.show()
```

```python
gray_img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
mtp.imshow(cv2.cvtColor(gray_img, cv2.COLOR_BGR2RGB))
mtp.axis('off')
mtp.show()
```

```python
inv_img = 255 - gray_img
mtp.imshow(cv2.cvtColor(inv_img, cv2.COLOR_BGR2RGB))
mtp.axis('off')
mtp.show()
```

```python
blurred = cv2.GaussianBlur(inv_img, (21,21), 0)
```

```python
inv_blur = 255 - blurred
pencil_sketch = cv2.divide(gray_img, inv_blur, scale=256.0)
mtp.imshow(cv2.cvtColor(pencil_sketch, cv2.COLOR_BGR2RGB))
mtp.axis('off')
mtp.show()
```
